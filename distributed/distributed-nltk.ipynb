
then, I started `dcenter` and `dworker 127.0.0.1:8787` (multiple times) on my local machine.

````python
from distributed import Executor
````

````python
e = Executor('127.0.0.1:8787')
````

````python
from glob import glob
filenames = sorted(glob('/Users/koverholt/nltk_data/corpora/state_union/*'))
````

````python
len(filenames)
````

````python
filenames[0:5]
````

````python
def read_data(filename):
    with open(filename) as f:
        lines = f.readlines()
    return lines
````

````python
read_data(filenames[0])
````

````python
result_1a = e.map(read_data, filenames)
````

````python
# Some of the tasks fail due to utf-8 formatting errors, we ignore the failed tasks
result_1a = [x for x in result_1a if x.status == 'finished']
````

````python
result_1a[0:5]
````

````python
%time len(result_1a[1].result())
````

````python
result_1b = e.submit(len, result_1a[1])
````

````python
%time result_1b.result()
````

````python
import nltk
````

````python
for line in read_data(filenames[1]):
    print(nltk.word_tokenize(line))
````

````python
def tokenize(lines):
    from toolz import concat
    
    return list(concat(map(nltk.word_tokenize, lines)))

tokenize(result_1a[0].result())
````

````python
result_2 = e.map(tokenize, result_1a)
````

````python
result_2[0:5]
````

````python
def pos_tag(words):
    return list(map(nltk.pos_tag, words))
````

````python
result_3 = e.map(nltk.pos_tag, result_2)
````

````python
from distributed import progress
````

````python
progress(result_3)
````

````python
e.gather(result_3)
````

````python
import dask.bag as db
b = db.from_filenames(filenames)
````
